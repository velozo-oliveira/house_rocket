```python
import pandas as pd 
data = pd.read_csv('DataSets/kc_house_data.csv/kc_house_data.csv')
from geopy.geocoders import Nominatim
```

```python


# Creating empty rows

data['road'] = "NA"
data['house_number'] = "NA"
data['neighbourhood'] = "NA"
data['city'] = "NA"
data['country'] = "NA"
data['state'] = "NA"

# Initialize Nominatim API
geolocator = Nominatim (user_agent='geoapiExercises')

i = 0
for i in range(len(data)):
    print('Loop: {} / {}'.format(i,len(data)))
    #make query
    query = str(data.loc[i,'lat']) + ',' + str(data.loc[i,'long'])

    # API Request
    response = geolocator.reverse (query)

    if 'house_number' in response.raw['address']:
         data.loc[i,'house_number']  = response.raw['address']['house_number']
        
    if 'road' in response.raw['address']:
        data.loc[i,'road'] = response.raw['address']['road']
    
    if 'neighbourhood' in response.raw['address']:
        data.loc[i,'neighbourhood'] = response.raw['address']['neighbourhood']
    
    if 'city' in response.raw['address']:
        data.loc[i,'city']          = response.raw['address']['city']
    
    if 'country' in response.raw['address']:
        data.loc[i,'country']       = response.raw['address']['country']
    
    if 'state' in response.raw['address']:
        data.loc[i,'state']         = response.raw['address']['state']

# print(response.raw['address']['road'])
# print(response.raw['address']['house_number'])
# print(response.raw['address']['neighbourhood'])
# print(response.raw['address']['city'])
# print(response.raw['address']['country'])
# print(response.raw['address']['state'])
```

# Filtros iterativos no mapa

```python
import plotly.express as px

data = pd.read_csv('DataSets/kc_house_data.csv/kc_house_data.csv')
houses = data[['id','lat','long','price']].copy()

#define price
for i in range (len(houses)):
    if houses.loc[i,'price'] <= 321950:
        houses.loc[i,'level'] = 0
        
    elif (houses.loc[i,'price'] > 321950) & (houses.loc[i,'price'] <= 450000):
        houses.loc[i,'level'] = 1
    
    elif (houses.loc[i,'price'] > 450000) & (houses.loc[i,'price'] <= 645000):
        houses.loc[i,'level'] = 2
        
    else: 
        houses.loc[i,'level'] = 3  

houses['level'] = houses['level'].astype(int)

fig = px.scatter_mapbox(houses,
                       lat = 'lat',
                       lon = 'long',
                       size= 'price',
                       color='level',
                       color_continuous_scale=px.colors.cyclical.IceFire,
                       size_max=15,
                       zoom=10)

fig.update_layout(mapbox_style='open-street-map')
fig.update_layout(height=600, margin = {'t':0,'l':0,'r':0,'b':0})
fig.show()
```

# Adicionando Filtros Iterativos 

```python
import ipywidgets as widgets
from ipywidgets import fixed
import pandas as pd 
import plotly.express as px 

df = pd.read_csv('DataSets/kc_house_data.csv/kc_house_data.csv')
houses = df[['id','lat','long','price']].copy()


df['is_waterfront'] = df['waterfront'].apply(lambda x: 'yes' if x==1 else 'no')

df['level'] = df['price'].apply(lambda x: 0 if x < 321950 else
                               1 if (x >= 321950) & (x < 450000) else
                               2 if (x >= 450000) & (x < 645000) else
                               3 )

df['level'] = df['level'].astype(int)
style = {'description_width':'initial'}

# Iterative buttons
price_limit = widgets.IntSlider( 
    value = 540000,
    min = 75000,
    max = 7700000,
    step = 1,
    description='Maximum Price',
    disable = False,
    style = style)

waterfront_bar = widgets.Dropdown(
    options = df['is_waterfront'].unique().tolist(),
    value = 'yes',
    description = 'Water View',
    disable = False)

def update_map (df,waterfront,price_limit):
    houses = df[(df['price'] <= price_limit) &
                (df['is_waterfront'] == waterfront)]
    [['id','lat','long','price','level']]
    
    fig = px.scatter_mapbox (houses,
                            lat='lat',
                            lon='long',
                            color='level',
                            size='price',
                            color_continuous_scale=px.colors.cyclical.IceFire,
                            size_max=15,
                            zoom=10)
    fig.update_layout(mapbox_style='open-street-map')
    fig.update_layout(height = 600, margin={'r':0,'t':0,'l':0,'b':0})
    fig.show()
```

```python
widgets.interactive(update_map,df=fixed(df),
                    waterfront=waterfront_bar,
                    price_limit=price_limit)
```

# Iteratividade com o dashboard

```python
import pandas as pd
import ipywidgets as widgets
from matplotlib import gridspec
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('DataSets/kc_house_data.csv/kc_house_data.csv')

data['year'] = pd.to_datetime(data['date']).dt.strftime('%Y')
data['date'] = pd.to_datetime(data['date']).dt.strftime('%Y-%m-%d')
data['year_week'] = pd.to_datetime(data['date']).dt.strftime('%Y-%U')

# Widgets to control data
date_limit = widgets.SelectionSlider(
    options = data['date'].sort_values().unique().tolist(),
    value = '2014-12-01',
    description = 'Disponivel',
    continuous_update=False,
    orientation='horizontal',
    readout=True)

def update_map (data,limit):
    #Filter Data
    df = data[data['date'] >= limit].copy()
    
    fig = plt.figure(figsize=(21,12))
    specs = gridspec.GridSpec(nrows = 2, ncols= 2, figure=fig)
    
    ax1 = fig.add_subplot(specs[0,:])
    ax2 = fig.add_subplot(specs[1,0])
    ax3 = fig.add_subplot(specs[1,1])
    
    by_year = df[['price','year']].groupby('year').sum().reset_index()
    ax1.bar(by_year['year'],by_year['price'])
    
    by_date = df[['price','date']].groupby('date').mean().reset_index()
    ax2.plot(by_date['date'],by_date['price'])
    ax2.set_title('Title: Average Price by date')
    
    
    by_year_week = df[['price','year_week']].groupby('year_week').mean().reset_index()
    ax3.bar(by_year_week['year_week'],by_year_week['price'])
    ax3.set_title('Average Price of Week of Year')
    plt.xticks(rotation=60);
    
    
    
    
    
```

```python
widgets.interactive(update_map,data=fixed(data), limit = date_limit)
```

```python
import pandas as pd 
import ipywidgets as widgets
from ipywidgets import fixed 
import plotly.express as ptx

data = pd.read_csv('DataSets/kc_house_data.csv/kc_house_data.csv')

style = {'description_width':'initial'}


min_living_room_size = widgets.FloatSlider( value=0,
                                            min=0,
                                            max=13540,
                                            step=5,
                                            description='Minimum living room size in sqft:',
                                            disabled=False,
                                            continuous_update=False,
                                            orientation='horizontal',
                                            readout=True,
                                            )

min_bathroom_number = widgets.FloatSlider(  value=0,
                                            min=data['bathrooms'].min(),
                                            max=data['bathrooms'].max(),
                                            step=0.5,
                                            description='Minimum bathroom number:',
                                            disabled=False,
                                            continuous_update=False,
                                            orientation='horizontal',
                                            readout=True,
                                            readout_format='.1f',)

price_limit = widgets.IntSlider(            value = 75000,
                                            min = 75000,
                                            max = 7700000,
                                            step = 1000,
                                            description='Maximum Price',
                                            disable = False,
                                            style = style)


max_sqft_basement_size = widgets.FloatSlider(value=0,
                                            min=data['sqft_basement'].min(),
                                            max=data['sqft_basement'].max(),
                                            step=5,
                                            description='Maximum basement size in sqft:',
                                            disabled=False,
                                            continuous_update=False,
                                            orientation='horizontal',
                                            readout=True,
                                            )

data['condition'] = data['condition'].astype(str)
house_condition = widgets.Dropdown(         options=['1','2','3','4','5'],
                                            value='1',
                                            description='Condition:',
                                            disabled=False,)

year_built = widgets.IntSlider(             value = 1970,
                                            min = 1950,
                                            max = 2020,
                                            step = 5,
                                            description='Year Built:',
                                            disable = False,
                                            style = style)



def update_map (data,living_room, bathroom_number, limit,basement_size,conditions,built):

    df = data[
    (data['sqft_living'] > living_room) & 
    (data['bathrooms'] > bathroom_number) & 
    (data['price'] < limit) & 
    (data['sqft_basement'] < basement_size) & 
    (data['condition'] == conditions) &
    (data['yr_built'] == built)
    ]
    data[['price','lat','long','sqft_living','bathrooms','condition','yr_built','sqft_basement']].copy()

    fig = ptx.scatter_mapbox(df,
                            lat='lat',
                            lon='long',
                            color_continuous_scale=ptx.colors.cyclical.IceFire,
                            size_max=15,
                            zoom=10)
    fig.update_layout(mapbox_style='open-street-map')
    fig.update_layout(height = 600, margin={'r':0,'t':0,'l':0,'b':0})
    fig.show()
```

```python
 widgets.interactive(update_map,
                     data = fixed(data), 
                     living_room = min_living_room_size, 
                     bathroom_number = min_bathroom_number,
                     limit = price_limit,
                     basement_size = max_sqft_basement_size,
                     conditions = house_condition,
                     built = year_built
                     )
```

```python
#5. Adicione os seguintes filtros no Dashboard
#  - Filtro por data disponivel para compra
#  - Filtro por ano de renovacao
#  - Filtro se possui vista para agua ou nao
```

```python
import pandas as pd
import ipywidgets as widgets
from ipywidgets import fixed
from matplotlib import gridspec
from matplotlib import pyplot as plt

data = pd.read_csv('DataSets/kc_house_data.csv/kc_house_data.csv')

data['date'] = pd.to_datetime(data['date']).dt.strftime('%m-%d-%Y')

data['is_waterfront'] = "NA"
i=0

for i in range(len(data)):
    
    if data.loc[i,'waterfront'] == 0:
        data.loc[i,'is_waterfront'] = 'No'
    
    if data.loc[i,'waterfront'] == 1:
        data.loc[i,'is_waterfront'] = 'Yes'


        


# Widgets to control data

waterfront_view = widgets.Dropdown(         options=['Yes','No'],
                                            value='Yes',
                                            description='Waterfront View:',
                                            disabled=False,)


date_limit = widgets.SelectionSlider(       options = data['date'].sort_values(ascending = False).unique().tolist(),
                                            value = '12-01-2014',
                                            description = 'Available:',
                                            continuous_update=False,
                                            orientation='horizontal',
                                            readout=True)


renovation_limit = widgets.SelectionSlider( options = data['yr_renovated'].sort_values().unique().tolist() ,
                                            value = 0,
                                            description = 'Year Renovated:',
                                            continuous_update=False,
                                            orientation='horizontal',
                                            readout=True)

def update_map (data,waterfront,limit,renovation):
    
    df = data[(data['is_waterfront'] == waterfront) &
              (data['date'] >= limit) &
              (data['yr_renovated'] == renovation)][['price','yr_renovated','is_waterfront','date']].copy()
    
    fig = plt.figure(figsize=(10,12))
    specs = gridspec.GridSpec(ncols=2,nrows = 2,figure=fig)
    
    ax1 = fig.add_subplot(specs[0,:])
    ax2 = fig.add_subplot(specs[1,0])
    ax3 = fig.add_subplot(specs[1,1])
    
    by_yr_renovated = df[['price', 'yr_renovated']].groupby('yr_renovated').mean().reset_index()
    ax1.bar(by_yr_renovated['yr_renovated'],by_yr_renovated['price'])
    ax1.set_title('Title: Average Price by Year Renovated')
    
    by_waterfront_view = df[['is_waterfront','price']].groupby('is_waterfront').mean().reset_index()
    ax2.bar(by_waterfront_view['is_waterfront'],by_waterfront_view['price'])
    ax2.set_title('Title: Average Price by Waterfront View')
    
    by_date_limit = df[['date','price']].groupby('date').mean().reset_index()
    ax3.bar(by_date_limit['date'],by_date_limit['price'])
    ax3.set_title('Title: Average Price by Date')
    plt.xticks(rotation=60)
    

```

```python
widgets.interactive(update_map,data = fixed(data), waterfront = waterfront_view, limit = date_limit, renovation = renovation_limit)
```
